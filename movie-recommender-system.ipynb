```python

import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
        
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python

def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
        break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

## Applying Transformer

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ",'') for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ",'') for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ",'') for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ",'') for i in x])
```

### Doing Text Vectorization
 We have to calculate the similarity score of movies or text..so for doing this we have to convert text into vectors

Whie doing this we will not consider "stopwors"(stopwords are the words that are used for sentence formation but they have 
not any contribution in sentence meaning...e.g are,for,and,of,do,from)

 Except stopwords we will do vectorization on theremaining text
 We have one class countvectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

so here we are getting same type of words like action,actions and actor,actors etc.
So we will apply stammen (stamen will replace same types of words into root word) e.g. for the words love,lovable,loved,
lovely..root word will be love


```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
##Example
ps.stem("loved")
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

We  have 4806 movies it means we have 4806 vectors and in evry vector we have 5000 numbers
we have to calucalate the distance of every movie with every other movie
Here we have to calculate Cosine Distance i.e. we will calculate the angle between the vectors.
if the distance is large then the similarity between movies is less i.e. Distance is inversely proportional to Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

here we are getting diagonal value as 1 since the similarity score of movies with itself is always 1

 We are creating a function such that if user will type any movie then the function will return 5 similar movies to the user

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]  ## Creating distances with index position and fetching out 5 top movies
```

```python
def recommend(movie):
    movie_index= new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6] 
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
  
```

```python
recommend('Batman Begins')
```

```python
recommend('Avatar')
```

```python
import pandas as pd
data = pd.read_csv("https://raw.githubusercontent.com/sunnysavita10/Naive-Bayes/main/SpamClassifier-with-ML/sms_spam_data/SMSSpamCollection.csv",sep = "\t",header = None,names=["label","messages"])
```

```python
import nltk   ##natural language toolkit
```

```python
import re ##regular expression (re) is used for finding out pattern in text dataset
```

```python
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
```

```python
from nltk.stem.porter import PorterStemmer  ##Stemming-getting a root word , e.g for the words likes,liked,likely,liking root word is like
```

```python
ps=PorterStemmer()
```

```python
corpus = []
for i in range(0,len(data)):
  review=re.sub('[^a-zA-Z]',' ',data['messages'][i])
  review=review.lower()
  review=review.split()

  review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
  review=' '.join(review)
  corpus.append(review)
```

```python
print(corpus)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=2500)
```

```python
X=cv.fit_transform(corpus).toarray()
```

```python
data['label']
```

```python
y = pd.get_dummies(data['label'],drop_first=True)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model2=MultinomialNB()
```

```python
model2.fit(X_train,y_train)
```

```python
y_pred2=model2.predict(X_test)
```

```python
accuracy_score(y_test,y_pred2)
```
